## Dos redes científicas, dos arquitecturas distintas

En esta sesión trabajaremos con dos redes de coautoría científica obtenidas desde Dimensions.

| Red       | Nodos | Enlaces |
| --------- | ----- | ------- |
| neuralNet | 1495  | 1451    |
| covidNet  | 13399 | 87524   |

Cada nodo representa un autor, y cada enlace representa una relación de coautoría entre dos autores.


````python
import networkx as nx
import requests
from io import BytesIO

urlNeural = "https://raw.githubusercontent.com/maestriaInnovacion-magallanes-redes/comunidades/main/data/neuralNet.gml"

urlCovid = "https://raw.githubusercontent.com/maestriaInnovacion-magallanes-redes/comunidades/main/data/covidNet.gml"

# neural network
responseNeural = requests.get(urlNeural)
neuralNet = nx.read_gml(BytesIO(responseNeural.content))

# covid collaboration network
responseCovid = requests.get(urlCovid)
covidNet = nx.read_gml(BytesIO(responseCovid.content))

print(neuralNet)
print(covidNet)
````

**Las redes dependen de cómo fueron construidas**

Las redes no existen de manera neutral.
Toda red depende de:

* los filtros utilizados,
* el periodo temporal,
* las palabras de búsqueda,
* y los datos disponibles.

Por ello, las propiedades observadas dependen parcialmente de las decisiones metodológicas tomadas durante la recolección.

En este caso:

> **neuralNet** fue construida usando:
> * artículos,
> * ingeniería,
> * año 2023,
> * open access bronce,
> * búsqueda “neural networks”.


> **covidNet** fue construida usando:    
> * artículos,
> * open access verde,
> * ingeniería,
> * búsqueda “covid”.


# Conectividad

Un componente representa un conjunto de nodos conectados entre sí mediante algún camino.

````python
nx.number_connected_components(neuralNet), nx.number_connected_components(covidNet)
````

<div class="alert alert-danger">

Los componentes <strong>NO SON COMUNIDADES</strong>, 
sino fragmentos completamente desconectados de la red.

</div>

> ¿Qué implica para la difusión de conocimiento que existan miles de componentes en una red de co-autores?

A primera vista ambas redes parecen muy fragmentadas.

Sin embargo, el tamaño del componente más grande cambia completamente la interpretación.


````python
max([len(c) for c in nx.connected_components(neuralNet)]),\
max([len(c) for c in nx.connected_components(covidNet)])
````

| Red       | Tamaño componente gigante |
| --------- | ------------------------- |
| neuralNet | 18                        |
| covidNet  | 1402                      |

**neuralNet** presenta:

* muchísimos componentes pequeños,
* baja conectividad global,
* y ausencia de un componente gigante dominante.

Esto sugiere una red altamente fragmentada, donde gran parte de los autores no se encuentran conectados indirectamente entre sí.

> La estructura observada limita potencialmente la alcanzabilidad global dentro de la red.

**covidNet** presenta una estructura distinta:

* un componente gigante considerablemente mayor,
* mayor conectividad indirecta,
* y una integración estructural mucho más amplia.

Esto sugiere que una gran cantidad de autores sí pertenece a una misma estructura conectada.

> La existencia de un componente gigante aumenta potencialmente la capacidad de difusión y circulación indirecta dentro de la red.


## Componente Gigante

<div class="alert alert-info">

Muchas métricas estructurales, como distancia promedio, diámetro o cercanía, 
requieren que exista conectividad entre nodos. 

Por ello, es común realizar estos análisis sobre el componente gigante.

</div>

````python
maxNeural = max(
    nx.connected_components(neuralNet),
    key=len
)

maxCovid = max(
    nx.connected_components(covidNet),
    key=len
)

neuralGigante = neuralNet.subgraph(maxNeural)

covidGigante = covidNet.subgraph(maxCovid)
````

````python
print(neuralGigante)
print(covidGigante)
````

````python
# =========================
# FUNCION RESUMEN
# =========================

# Función para métricas globales del componente gigante
def metricas_gc(G, nombre):
    return {
        "Red": nombre,
        "Nodos": G.number_of_nodes(),
        "Enlaces": G.number_of_edges(),
        "Diámetro": nx.diameter(G),
        "Distancia Media": nx.average_shortest_path_length(G),
        "Densidad": nx.density(G)
    }

# Tabla
tabla_gc = pd.DataFrame([
    metricas_gc(neuralGigante, "neuralGigante"),
    metricas_gc(covidGigante, "covidGigante")
])

pd.options.display.float_format = "{:.4f}".format

tabla_gc
````

Las diferencias entre ambos componentes gigantes muestran arquitecturas estructurales profundamente distintas. El componente gigante de **neuralNet** corresponde a una estructura completamente conectada, donde todos los autores mantienen vínculos directos entre sí, produciendo máxima densidad y cohesión estructural.

````python
nx.draw(neuralGigante)
````

Sin embargo, esta cohesión ocurre dentro de un grupo muy pequeño, lo que contrasta con la fuerte fragmentación observada en la red completa. Por el contrario, **covidNet** presenta un componente gigante mucho más amplio e integrado, capaz de conectar indirectamente a más de mil autores, aunque con menor densidad, menor cohesión estructural y mayores distancias promedio. Esto sugiere dos configuraciones estructurales distintas: una basada en grupos pequeños altamente cohesionados y otra basada en una infraestructura de colaboración extensa con cohesión más débil.


# Comunidades



El componente gigante de **neuralNet** no presenta suficiente complejidad estructural interna para realizar un análisis significativo de comunidades. Debido a que todos los nodos se encuentran directamente conectados entre sí, no existen agrupamientos internos, puentes ni regiones periféricas que permitan identificar subestructuras diferenciadas. Por ello, el análisis de comunidades resulta mucho más relevante en redes extensas y menos cohesionadas, como ocurre en el caso de **covidNet**.


Hasta ahora hemos analizado propiedades globales de las redes:

* densidad,
* distancia media,
* diámetro,
* alcanzabilidad.

Sin embargo, muchas redes sociales no se organizan homogéneamente.
Frecuentemente aparecen:

* grupos densos,
* zonas periféricas,
* regiones altamente conectadas,
* y puentes entre grupos.

Para entender esa organización interna analizaremos:

1. clustering coefficient,
2. transitivity,
3. cliques,
4. comunidades.

---

## 1. Clustering coefficient

El clustering coefficient mide hasta qué punto los vecinos de un nodo también se encuentran conectados entre sí.

En términos simples:

> ¿Los amigos de mis amigos también son amigos entre sí?

Cuando el clustering es alto:

* aparecen triángulos,
* la cohesión local aumenta,
* y la red tiende a formar grupos densos.

En la figura 3, el nodo naranja mantiene cuatro vecinos (2, 3, 4 y 5). Entre esos vecinos podrían existir hasta seis conexiones posibles. A medida que aumentan las conexiones entre ellos (líneas rojas), el clustering coefficient del nodo aumenta progresivamente desde 0 hasta 1.

<figure>
<img src="https://github.com/SocialAnalytics-StrategicIntelligence/SocialNet_analytics/raw/main/picsForSession/LocalClustCoeff.png" style="width:70%">
<figcaption align = "center"> Figure 3. Average clustering coefficient. 
 </figcaption>
 <a href="https://www.researchgate.net/publication/42763973_Toward_a_Network_Perspective_of_the_Study_of_Resilience_in_Social-Ecological_Systems">Source: Ecology and Society</a>
</figure>

Cuando el valor es:

* cercano a 0, los vecinos casi no se conectan entre sí;
* cercano a 1, los vecinos forman un grupo altamente cohesionado.

Por ello, el clustering coefficient es una medida de cohesión local y cierre estructural.


````python
nx.average_clustering(covidGigante), nx.average_clustering(neuralGigante)
````

El clustering coefficient promedio muestra una diferencia importante entre cohesión global y cohesión local. Aunque **covidGigante** presenta una densidad muy baja (~ 0.0084), su clustering coefficient promedio es extremadamente alto (~ 0.92). Esto indica que, aun cuando la red es globalmente dispersa, los vecinos de cada nodo tienden fuertemente a conectarse entre sí, formando estructuras locales altamente cohesionadas. Nuestra red puede es simultáneamente extensa y dispersa a nivel global, pero altamente cohesionada a escala local.


## 2. Transitivity

La transitivity mide la proporción global de triángulos observados en la red.

Mientras el clustering promedio promedia nodos individualmente, la transitivity resume la tendencia global de la red a formar triángulos.



````python
nx.transitivity(covidGigante)
````

````python


```

---

## Interpretación

Valores altos indican:

* mayor cierre estructural,
* mayor cohesión global,
* y mayor presencia de relaciones triangulares.

---

# Diferencia conceptual

| Métrica                | Nivel  |
| ---------------------- | ------ |
| Clustering coefficient | Local  |
| Transitivity           | Global |

---

# 3. Cliques

Una clique es un subconjunto de nodos donde todos se encuentran conectados directamente entre sí.

Representan el máximo nivel posible de cohesión estructural.

---

## Buscar cliques

```python
cliquesCovid = list(nx.find_cliques(gcCovid))
```

---

## Tamaño de las mayores cliques

```python
sorted(
    [len(c) for c in cliquesCovid],
    reverse=True
)[:10]
```

---

## Interpretación

Las cliques permiten identificar:

* núcleos extremadamente cohesionados,
* grupos densos,
* y posibles regiones de colaboración intensa.

Sin embargo:

* las cliques son estructuras muy estrictas,
* y grandes cliques son relativamente raras en redes reales extensas.

---

# 4. Comunidades

Las comunidades representan grupos de nodos más densamente conectados entre sí que con el resto de la red.

A diferencia de las cliques:

* no requieren conexión total,
* sino cohesión relativa.

---

# Detección de comunidades

Usaremos greedy modularity.

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
comunidadesCovid = greedy_modularity_communities(gcCovid)
```

---

## Número de comunidades

```python
len(comunidadesCovid)
```

---

## Tamaño de comunidades

```python
sorted(
    [len(c) for c in comunidadesCovid],
    reverse=True
)[:10]
```

---

# Modularidad

La modularidad mide qué tan separadas se encuentran las comunidades observadas respecto de una red aleatoria equivalente.

Valores altos indican:

* fuerte estructura comunitaria,
* mayor separación entre grupos,
* y cohesión interna relativamente alta.

---

## Cálculo de modularidad

```python
from networkx.algorithms.community.quality import modularity

modularity(
    gcCovid,
    comunidadesCovid
)
```

---

# Interpretación general

Las comunidades permiten identificar:

* agrupamientos estructurales,
* posibles subcampos,
* regiones de colaboración,
* y patrones internos de organización.

A diferencia de los componentes:

* las comunidades sí pertenecen a una misma estructura conectada,
* pero presentan cohesión diferencial dentro de la red.
````

````python



---

# Componente gigante



---

# Interpretación de neuralNet



---

# Componentes NO son comunidades

Los componentes representan:

* desconexión absoluta.

Las comunidades representan:

* agrupamientos internos dentro de una red conectada.

Por ello:

* primero analizamos componentes,
* luego comunidades dentro del componente gigante.

---

# Hacia los modelos de red

Las diferencias observadas pueden interpretarse usando modelos clásicos de redes:

| Modelo          | Rasgo principal                         |
| --------------- | --------------------------------------- |
| Erdös-Rényi     | conexiones relativamente homogéneas     |
| Barabási-Albert | hubs altamente conectados               |
| Watts-Strogatz  | comunidades conectadas mediante puentes |

Las redes reales rara vez corresponden exactamente a un modelo puro.

Sin embargo, los modelos ayudan a interpretar:

* mecanismos de colaboración,
* difusión de conocimiento,
* fragmentación,
* y arquitectura estructural.

---

# Idea central

Las redes científicas no solo representan relaciones entre autores.

También reflejan:

* formas de organización social,
* mecanismos de colaboración,
* estructuras de difusión,
* y condiciones de producción de conocimiento.

Por ello, analizar la estructura de una red permite comprender parcialmente cómo circula el conocimiento dentro de sistemas sociales complejos.
````

## Exploration of network communities

* **Average clustering coefficient**: is the average of the local clustering coefficients of the nodes. If all the neighbors of every node are connected among one another, you get 1; if none of them are connected you get zero. 




````python
# count_zeros=False -> to make results compatible with R!
nx.average_clustering(topsNet_giant,count_zeros=False)
````

* **Transitivity**: Tendency of local clustering.

````python
# Probability that two tech guys with a common colleague, are also colleagues.
nx.transitivity(topsNet_giant)
````

* **Assortativity (degree)**: it is a measure to see if nodes are connecting to other nodes similar in degree.  Closer to 1 means higher assortativity, closer to -1 diassortativity; while 0 is no assortitivity.

````python
nx.degree_assortativity_coefficient(topsNet_giant)
````

You can also compute assortativity using an attribute of interest:

````python
nx.attribute_assortativity_coefficient(topsNet_giant,'male') # positive if tend to connect with same sex.
````

````python
nx.attribute_assortativity_coefficient(topsNet_giant,'popularity') # positive if tend to connect with same sex.
````

If a network presents cliques makes you suspect that there can be **communities**.

This is a huge field of research, let me me show you one of the algorithms known as the Louvain method.



````python
nx.draw_kamada_kawai(topsNet_giant)
````

````python
# Calculating the betweenness centrality
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant)
list(btw_centrality.items())[:4]
````

````python


# Sorting based on the betweenness centraliy and displaying the first 10 edges.
sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0:10]
````

````python
bye_edge_1 = sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]
bye_edge_1
````

````python
pos=nx.kamada_kawai_layout(topsNet_giant)
nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0]],width=6,edge_color='red' );
````

````python
topsNet_giant_1=topsNet_giant.copy()
topsNet_giant_1.remove_edge(*bye_edge_1[0])
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant_1)
bye_edge_2=sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]

nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0],bye_edge_2[0]],width=6,edge_color='red' );
````

````python
topsNet_giant_2=topsNet_giant_1.copy()
topsNet_giant_2.remove_edge(*bye_edge_2[0])
btw_centrality = nx.algorithms.centrality.edge_betweenness_centrality(topsNet_giant_2)
bye_edge_3=sorted(btw_centrality.items(), key = lambda item:item[1], reverse = True)[0]

nx.draw(topsNet_giant,pos=pos, width=0.1,edge_color='grey')
nx.draw_networkx_edges(topsNet_giant,pos=pos,edgelist=[bye_edge_1[0],bye_edge_2[0],bye_edge_3[0]],width=6,edge_color='red' );
````

````python
topsGirvNew = nx.algorithms.community.centrality.girvan_newman(topsNet_giant)
topsGirvNew_result=[x for x in topsGirvNew if len(x)==4][0]
````

````python
girvanNewman_attr={z:x for x,y  in enumerate(topsGirvNew_result) for z in y }
girvanNewman_attr
````

````python
nx.set_node_attributes(topsNet_giant, girvanNewman_attr,'girvanNewman')
````

````python
list(dict(topsNet_giant.nodes(data=True)).items())[:4]
````

````python
pos=nx.spring_layout(topsNet_giant,k=3) 
plt.figure(figsize=(8,8))
color_map = plt.get_cmap("cool")
valuesForColors=[n[1]['girvanNewman'] for n in topsNet_giant.nodes(data=True)]
nx.draw(topsNet_giant,pos=pos,node_color=valuesForColors,cmap=color_map,with_labels=True,edge_color='lightblue')
````

````python
topsLouvain=nx.community.louvain_communities(topsNet_giant, seed=123)
louvain_attr={z:x for x,y  in enumerate(topsLouvain) for z in y }
louvain_attr
````

````python
topsLouvain=nx.community.louvain_communities(topsNet_giant, seed=123)
louvain_attr={z:x for x,y  in enumerate(topsLouvain) for z in y }
nx.set_node_attributes(topsNet_giant, louvain_attr,'louvain')

pos=nx.spring_layout(topsNet_giant,k=3) 
plt.figure(figsize=(8,8))
color_map = plt.get_cmap("cool")
valuesForColors=[n[1]['louvain'] for n in topsNet_giant.nodes(data=True)]
nx.draw(topsNet_giant,pos=pos,node_color=valuesForColors,cmap=color_map,with_labels=True,edge_color='lightblue')
````

````python
communities = list(nx.community.girvan_newman(topsNet_giant))
modularity_df = pd.DataFrame(
    [
        [k + 1, nx.community.modularity(topsNet_giant, communities[k])]
        for k in range(len(communities))
    ],
    columns=["k", "modularity"],
)


# Plot change in modularity as the important edges are removed
modularity_df.plot.bar(
    x="k",
    color="#F2D140",
    title="Modularity Trend for Girvan-Newman Community Detection",
)
````
